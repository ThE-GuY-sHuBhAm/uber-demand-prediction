```python
import numpy as np
import pandas as pd
import dask.dataframe as dd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# paths for the three dfs

df_jan_path = "../data/raw/yellow_tripdata_2016-01.csv"
df_feb_path = "../data/raw/yellow_tripdata_2016-02.csv"
df_mar_path = "../data/raw/yellow_tripdata_2016-03.csv"

# load the dataframes

df_jan = dd.read_csv(df_jan_path, assume_missing=True)
df_feb = dd.read_csv(df_feb_path, assume_missing=True)
df_mar = dd.read_csv(df_mar_path, assume_missing=True)
```

```python
# concat the three dataframes as one

df_final = dd.concat([df_jan, df_feb, df_mar], axis=0).drop(columns=['extra', 'mta_tax', 'tip_amount', 'tolls_amount',
       'improvement_surcharge', 'total_amount', 'payment_type'])
```

```python
df_final['pickup_latitude'].min().compute()
df_final['pickup_latitude'].max().compute()
```

```python
Q1 = df_final['pickup_latitude'].quantile(0.25).compute()
Q3 = df_final['pickup_latitude'].quantile(0.75).compute()

# 2. Calculate IQR
IQR = Q3 - Q1

# 3. Define Upper and Lower Bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. Filter data
filtered_df = df_final[(df_final['pickup_latitude'] >= lower_bound) & (df_final['pickup_latitude'] <= upper_bound)]

# 5. Calculate Max
max_value = filtered_df['pickup_latitude'].max().compute()
min_value = filtered_df['pickup_latitude'].min().compute()
```
